```python
train = pd.read_csv("/kaggle/input/cs-3780-5780-how-do-you-feel/train.csv")
train_text = train["text"]
train_label = train["label"]

test = pd.read_csv("/kaggle/input/cs-3780-5780-how-do-you-feel/test.csv")
test_id = test["id"]
test_text = test["text"]
```

```python
#preprocessing 
#changing text data to bag of words
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
train_text.astype(str)
vectorizer = CountVectorizer()

train_text = vectorizer.fit_transform(train_text)
train_text = pd.DataFrame(train_text.toarray(), columns=vectorizer.get_feature_names_out())

# Show the transformed DataFrame
print(train_text)
```

```python
#train test validation split
from sklearn.model_selection import train_test_split

#changing the training dataset to numpy arrays
train_data = np.array(train_text)
label_data = np.array(train_label)

#splitting into test train and validation 

x_train, x_val, y_train, y_val= train_test_split(train_data, label_data, test_size= 0.3, random_state=42)
```

```python
print("the size of the train data is {},{}".format(len(x_train),len(y_train)))
print("the size of the validation data is{},{}".format(len(x_val),len(y_val)))
```

```python
#decisiom tree for classification of sentiment
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()

dtree1 = dtree.fit(x_train,y_train)
train_tree_pred = dtree1.predict(x_train)

#calculate the accuracy score
score = 0 
for i in range(len(x_train)):
    if train_tree_pred[i] == y_train[i]:
        score = score +1
    else:
        score = score +0
accuracy = score/len(y_train)
```
